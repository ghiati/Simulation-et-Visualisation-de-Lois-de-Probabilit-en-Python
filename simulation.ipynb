```python
import random
import math
import matplotlib.pyplot as plt
import probability_laws as pl


def visualiser_simulation(data, titre):
    plt.hist(data, bins=30, edgecolor='black')
    plt.title(titre)
    plt.xlabel('Valeur')
    plt.ylabel('Fr√©quence')
    plt.show()


taille = 1000

```

```python

# Bernoulli
p_bernoulli = 0.5
data_bernoulli = pl.bernoulli_simulation(p_bernoulli, taille)
visualiser_simulation(data_bernoulli, 'Loi de Bernoulli')

```

```python

# Binomiale
n_binomial = 10
p_binomial = 0.5
data_binomiale = pl.binomial_simulation(n_binomial, p_binomial, taille)
visualiser_simulation(data_binomiale, 'Loi Binomiale')

```

```python
# G√©om√©trique
p_geometric = 0.5
data_geometrique = pl.geometric_simulation(p_geometric, taille)
visualiser_simulation(data_geometrique, 'Loi G√©om√©trique')

```

```python
# Poisson
lam_poisson = 3
data_poisson = pl.poisson_simulation(lam_poisson, taille)
visualiser_simulation(data_poisson, 'Loi de Poisson')

```

```python
# Uniforme continue
a_uniforme = 0
b_uniforme = 1
data_uniforme = pl.uniform_simulation(a_uniforme, b_uniforme, taille)
visualiser_simulation(data_uniforme, 'Loi Uniforme Continue')

```

```python
# Normale
mu_normal = 0
sigma_normal = 1
data_normale = pl.normal_simulation(mu_normal, sigma_normal, taille)
visualiser_simulation(data_normale, 'Loi Normale')

```

```python
# Exponentielle
lam_exponential = 1
data_exponentielle = pl.exponential_simulation(lam_exponential, taille)
visualiser_simulation(data_exponentielle, 'Loi Exponentielle')
```
